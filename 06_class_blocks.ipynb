# How does a player's blocking ability relate to their year in school? 

### We need to pull game-by-game records of player's blocks and minutes to figure this out. We'll include a couple of clauses to look only at 2017 and exclude stats for which our original dataset doesn't have a player's class recorded.

```python
from pandas.io import gbq
project_id = '[YOUR_PROJECT_ID]'
```

```python
query = """
SELECT
  class,
  SUM(blocks) AS blocks,
  SUM(minutes_int64) AS minutes,
  SUM(blocks)/SUM(minutes_int64) AS blocks_per_minute
FROM
  `bigquery-public-data.ncaa_basketball.mbb_players_games_sr`
WHERE
  class != ""
  AND season = 2017
GROUP BY
  class
ORDER BY
  blocks_per_minute DESC
"""

class_blocks = gbq.read_gbq(query=query, dialect ='standard', project_id=project_id)
class_blocks
```
